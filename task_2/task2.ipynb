```python
import re
import nltk
from nltk.corpus import stopwords
import string
```

```python


nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_tweet(tweet):
    tweet = str(tweet)  # Ensure it's a string
    tweet = tweet.lower()  # Lowercase
    tweet = re.sub(r"http\S+|www\S+|pic.twitter\S+", "", tweet)  # Remove URLs
    tweet = re.sub(r"@\w+", "", tweet)  # Remove mentions
    tweet = re.sub(r"#\w+", "", tweet)  # Remove hashtags
    tweet = tweet.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    tweet = re.sub(r"\d+", "", tweet)  # Remove numbers
    tweet = re.sub(r"[^\x00-\x7F]+", "", tweet)  # Remove non-ASCII (like emojis)
    tweet = re.sub(r"\s+", " ", tweet)  # Remove extra spaces
    tweet = " ".join([word for word in tweet.split() if word not in stop_words])  # Remove stopwords
    return tweet.strip()


df4['Cleaned_Tweet'] = df4['Tweet Content'].apply(clean_tweet)
```

```python
df4.sample(25)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  
    
    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df4['Sentiment'] = df4['Cleaned_Tweet'].apply(get_sentiment)
```

## Sentiment analysis for overall data

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia = SentimentIntensityAnalyzer()

def analyze_sentiment_vader(text):
    score = sia.polarity_scores(text)
    compound = score['compound']
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Apply to the cleaned tweets
df4['VADER_Sentiment'] = df4['Cleaned_Tweet'].apply(analyze_sentiment_vader)
```

```python
df4['VADER_Compound'] = df4['Cleaned_Tweet'].apply(lambda x: sia.polarity_scores(x)['compound'])
```
